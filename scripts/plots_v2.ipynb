# Language dependency

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

y_axis = ['Independent', 'Dependent']
x_axis = [46.87, 53.13]

f = plt.figure()
f.set_figwidth(5)
f.set_figheight(2.5)

bars = plt.barh(y_axis, x_axis, color=(0.1, 0.1, 0.1, 0.1), edgecolor='black')
plt.ylabel('Language Dependency')
plt.xlabel('Value in percentage')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().invert_yaxis()

locator = ticker.MultipleLocator(2)
plt.gca().xaxis.set_major_locator(locator)
formatter = ticker.StrMethodFormatter("{x:.0f}")
plt.gca().xaxis.set_major_formatter(formatter)

plt.gca().xaxis.set_ticks(np.arange(1, max(x_axis) + 1, 9.0))

# Making the title bold and increasing font size
# plt.title('Language Dependency', fontweight='bold', fontsize=12)

# Adding data labels outside the bars
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}',
             va='center', color='black', fontsize=7,fontweight='bold')
plt.savefig('../figures/language_dependency.pdf', bbox_inches='tight', pad_inches = 0)
plt.show()
```
